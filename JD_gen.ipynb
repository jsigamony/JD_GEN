```python
%%writefile app.py
import streamlit as st
from transformers import BitsAndBytesConfig, AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

# Model setup
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16,
)

tokenizer = AutoTokenizer.from_pretrained("unsloth/gemma-2-2b-it")
model = AutoModelForCausalLM.from_pretrained("unsloth/gemma-2-2b-it", quantization_config=bnb_config)

llm_model = pipeline(
    model=model,
    tokenizer=tokenizer,
    task="text-generation",
    do_sample=True,
    temperature=0.1,
    repetition_penalty=1.1,
    return_full_text=False,
    max_new_tokens=200,
)

messages = [
    {"role": "user",
    "content": """I want you to generate a job description from the following information.
    Your generated job description should contain these headings : Job Title, Job Description, Job Responsibility, Qualifications and How to Apply.
    Under each heading, write bullet points.

    Information for you:

    Job Title : {job_title}
    Job Experience : {job_exp}
    Job Qualification : {job_qual}
    Skills : {skills}

    """},
]

prompt_template = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

def main():
    st.title("Job Description Generator")
    job_title = st.text_input("Enter the title of the job:")
    job_exp = st.text_input("Input the number of experience for this job:")
    job_qual = st.text_input("Input the required qualification for this job (e.g BS/MS in cs,ai,bba):")
    skills = st.text_input("Input the skills needed for this job in a comma-separated way (e.g python, c++, photoshop)")

    if st.button("Generate Job Description"):
        final_prompt = prompt_template.format(job_title=job_title, job_exp=job_exp, job_qual=job_qual, skills=skills)
        result = llm_model(final_prompt)
        st.write(result[0]['generated_text'])

if __name__ == "__main__":
    main()
```
